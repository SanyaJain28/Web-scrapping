```python
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
```

```python
from scrapy import Selector
```

```python
URL='https://www.walmart.com/reviews/product/247034695'
```

```python
#import the webpage content
r=requests.get(URL)
```

```python
#convert the page into a beautiful soup object
soup=bs(r.content,'html.parser')
```

```python
print(soup.prettify())
```

```python

header=soup.select("div h3")
header
```

```python
review=[]
for i in range(len(header)):
    review.append(header[i].get_text())
review
```

```python
date=soup.find_all("div", class_="review-date")
date
```

```python
dates=[]
for i in range(len(date)):
    dates.append(date[i].get_text())
dates
```

```python
text=soup.find_all("div", class_="review-text")
text
```

```python
reviewtext=[]
for i in range(len(text)):
    reviewtext.append(text[i].get_text())
reviewtext
```

```python
stars=soup.find_all("span",class_="elc-icon star star-small star-rated elc-icon-star-rating")
stars
```

```python
df=pd.DataFrame()
```

```python
df['Review Heading']=review
df['Date']=dates
df['Reviews']=reviewtext
df
```

```python
df.to_csv(r'C:\Users\Sanya Jain\Documents\Important docs\Walmartreviews.csv', index=True)
```
